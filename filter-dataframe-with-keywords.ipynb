# Filter dataframe with keywords using `str.find()`

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/Average_Daily_Traffic_Counts.csv')
```

### Find any street that includes the letters 'Hwy' in the street name
The `str.find()` method will return the position of the word in the string, so I need to specific > 0 to return only those records where a word is found.

```python
df[df.Street.str.find('Hwy')>0]
```
